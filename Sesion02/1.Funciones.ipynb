```python
names = [
    "  juan perez  ",
    "maria lopez",
    "JUAN PEREZ",
    "ana Torresㅤ",
    " ",
    "carlos ruiz",
    "ana torres",
    "   luis diaz"
]
```

```python
from typing import List
def normalizar_nombres(names: List[str]) -> List[str]:
    """
    Limpia y normaliza una lista de nombres
    -Elimina espacios que estan al inicio y al final del nombre
    -Elimina valores vacios
    """
    nombres_normalizados = [name.strip().title() for name in names if name.strip()]
    return nombres_normalizados
```

```python
normalizar = normalizar_nombres(names)
print(normalizar)
```

```python
def encuentra_duplicados(items: List[str]) -> List[str]:
    conjunto = set()
    duplicados = set()

    for item in items:
        if item in conjunto:
            duplicados.add(item)
        conjunto.add(item)
    
    return list(duplicados)
```

```python
duplicados = encuentra_duplicados(normalizar)
print(duplicados)
```

```python
def limpia_cadena(texto: str) -> str:
    """Elimina caracteres especiales
    """
    return ''.join(c for c in texto if c.isalnum() or c.isspace())
```

```python
texto = 'ana Torres '
texto_limpio = limpia_cadena(texto)
print([texto_limpio])
```

```python
from datetime import datetime

def validar_fecha(fecha_str: str) -> bool:

    try:
        datetime.strptime(fecha_str, "%d-%m-%Y")
        return True
    except Exception as e:
        print(e)
        return False
```

```python
validar_fecha("20-01-2020")
```
