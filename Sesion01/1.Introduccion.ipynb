```python
# Crea un programa que:
# 	1.	Pida números al usuario usando input().
# 	2.	Los guarde en una lista.
# 	3.	Termine cuando el usuario escriba "fin".
# 	4.	Muestre:
# 	•	La lista completa
# 	•	El número mayor
# 	•	El número menor
# 	•	El promedio
```

```python
nums = []

while True:
    s = input('Ingrese un numero o escriba fin ').lower()
    if s == 'fin':
        break
    
    n = float(s)
    nums.append(n)

if nums:
    promedio = sum(nums) / len(nums)
    print("lista: ", nums)
    print("Mayor: ", max(nums))
    print("Menor: ", min(nums))
    print("Promedio: ", promedio)
```

```python
# Simula un registro de alumnos:
# 	1.	Usa un diccionario donde:
# 	•	clave = nombre del alumno
# 	•	valor = nota final
# 	2.	Pide datos al usuario hasta que escriba "salir".
# 	3.	Luego muestra:
# 	•	Todos los alumnos
# 	•	El alumno con mayor nota
# 	•	El alumno con menor nota
```

```python
notas = {}

while True:
    nombre = input('Nombre o salir: ').lower()

    if nombre == 'salir':
        break

    nota = float(input('Nota: '))
    notas[nombre] = nota

if notas:
    print('Registro: ', notas)
    mayor = max(notas, key=notas.get)
    menor = min(notas, key=notas.get)
    print('Mayor: ', mayor)
    print('Menor: ', menor)
```
